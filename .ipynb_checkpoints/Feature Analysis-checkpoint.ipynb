```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('original_data.csv')
data.head()
```

```python
#list all categories
data.age.unique()
```

```python
#iterate the data by ages and store the number of different status in that categories into a dictionary
age_dic = {}
for index,row in data.iterrows():
    key = row['age']
    i = 0 if row['y'] == 'yes' else 1
    tmp = [0]*2
    if key in age_dic:
        tmp = age_dic[key]
    tmp[i] += 1
    age_dic[key] = tmp
```

```python
#convert the dictionary into dataframe for easy processing
age_df = pd.DataFrame.from_dict(age_dic)
age_df = age_df.append(dict(age_df.sum()),ignore_index=True).T
#sort the dataframe by the total amount of project
age_df = age_df.sort_index()
```

```python
#plot the stacked bar
N = len(age_df)
suc = age_df.iloc[:,0]
fai = age_df.iloc[:,1]
r = age_df[0]/age_df[2]

#separate the project of different status in differnet categories
ind = np.arange(N)
width = 0.35

fig, ax1 = plt.subplots(figsize=(20,20))
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

#setup the vertical bar
p1 = ax1.bar(ind, suc, width, color = 'g')
p2 = ax1.bar(ind, fai, width, bottom=np.array(suc), color = 'r')
p3 = ax2.plot(ind,r,color = 'b')


ax1.set_xlabel('Age',fontsize = 25)
ax1.set_ylabel('Result',fontsize = 25)
ax2.set_ylabel('Rate', fontsize = 25)

plt.title(' Stacked With Result',fontsize = 25)
plt.xticks(ind, age_df.index,fontsize = 20,rotation=90)
plt.yticks(fontsize = 25)
ax2 = plt.axes 

plt.legend((p1[0], p2[0], p3[0]), ('Yes', 'No','Rate'),loc='best',fontsize = 30)
plt.savefig('age_distribute_with_rate.png')
plt.show()
```
